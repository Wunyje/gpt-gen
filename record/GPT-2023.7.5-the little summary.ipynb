````python
import os

def create_directory_structure(base_dir, structure):
    for key, value in structure.items():
        path = os.path.join(base_dir, key)
        if isinstance(value, dict):
            os.makedirs(path, exist_ok=True)
            create_directory_structure(path, value)
        else:
            open(path, 'w').close()

client_structure = {
    'src': {
        'main.c': None,
        'pipeline.c': None,
        'protocols.c': None,
        'playback.c': None,
        'cli.c': None
    },
    'include': {
        'pipeline.h': None,
        'protocols.h': None,
        'playback.h': None,
        'cli.h': None
    },
    'bin': {},
    'data': {},
    'config': {},
    'Makefile': None,
    'README.md': None
}

server_structure = {
    'src': {
        'main.c': None,
        'pipeline.c': None,
        'protocols.c': None,
        'storage.c': None,
        'conversion.c': None,
        'bandwidth.c': None,
        'cli.c': None,
        'vod.c': None
    },
    'include': {
        'pipeline.h': None,
        'protocols.h': None,
        'storage.h': None,
        'conversion.h': None,
        'bandwidth.h': None,
        'cli.h': None,
        'vod.h': None
    },
    'bin': {},
    'data': {},
    'config': {},
    'Makefile': None,
    'README.md': None
}

create_directory_structure('client', client_structure)
create_directory_structure('server', server_structure)
````
